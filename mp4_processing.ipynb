```python
import pylab
import imageio
import numpy as np
import matplotlib.pyplot as plt
from skimage.measure import block_reduce as reduce
import scipy.ndimage
import skvideo.io
import time
```

```python
t1 = time.time()
vid1 = imageio.get_reader("sample1.mp4", "ffmpeg")
vid1_img = np.stack([vid1.get_data(i) for i in range(vid1.get_length())])
t2 = time.time()
t2-t1
```

```python
t1 = time.time()
vid1 = skvideo.io.vread("sample1.mp4")
t2 = time.time()
t2-t1
```
